```python
#---------CREATE BOARD---------#

def display_board(game_board):

    row1 = ' | '.join(str(i) for i in game_board[0:3])
    row2 = ' | '.join(str(i) for i in game_board[3:6])
    row3 = ' | '.join(str(i) for i in game_board[6:9])
    row_break = '---------'
    print('CURRENT MATCH')
    print(row1)
    print(row_break)
    print(row2)
    print(row_break)
    print(row3)
```

```python
#---------SELECT STARTING PLAYER---------#

from random import randint

def choose_first():
    player = randint(1, 2)

    print('You are player {}'.format(player))
    return player
```

```python
#---------SELECT MARKER TO PLAY WITH---------#

def select_marker():

    markers = ['X', 'O']
    player1 = ''
    player2 = ''
    
    while player1 not in markers:
        player1 = input('Select a marker for PLAYER 1: ')
        
        if player1 not in markers:
            print('Please select X or O to continue.')

    else:
        player2 = 'O' if player1 == 'X' else 'X'
        print('PLAYER 1 chose {}'.format(player1))
        print('PLAYER 2 chose {}'.format(player2))        
        return player1, player2, markers
```

```python
#---------WINNING CONDITIONS---------#

def win_condition(game_board):
    '''list_arr[row][col] if [['X', 'X']]'''
    game_over = False
    player = 1

    def win_cond(game_board):
        wind_cond_list = all(x == game_board[0] for x in game_board)

        return wind_cond_list, game_board

    def get_diagonal(game_board):
        diag = np.reshape(game_board, (3,3))
        diagonal_matrix = [r[i] for i, r in enumerate(diag)]
        reverse_diagonal_matrix = [r[~i] for i, r in enumerate(diag)]

        return diagonal_matrix, reverse_diagonal_matrix

    while game_over == False:
    
        for i in range(0, len(game_board), 3):
            win_cond_row, row_pattern = win_cond(game_board[i:i+3])
            win_cond_col, col_pattern = win_cond(list(np.concatenate(np.transpose(np.reshape(game_board, (3,3)))))[i:i+3]) 

            if win_cond_row == True:
                print('WIN >>> MATCHING MARKERS >>> ROW')
                print('Player {} wins'.format(player))
                print(row_pattern)
                
                return True

            elif win_cond_col == True:
                print('WIN >>> MATCHING MARKERS >>> COLUMN')
                print('Player {} wins'.format(player))
                print(col_pattern)
                
                return True

        diagonal_matrix, reverse_diagonal_matrix = get_diagonal(game_board)
        win_cond_diag, diagonal_matrix = win_cond(diagonal_matrix)
        win_cond_rev_diag, reverse_diagonal_matrix = win_cond(reverse_diagonal_matrix)
        
        if win_cond_diag == True:
            print('WIN >>> MATCHING MARKERS >>> DIAGONAL')
            print('Player {} wins'.format(player))
            print(diagonal_matrix)
            
            return True

        elif win_cond_rev_diag == True:
            print('WIN >>> MATCHING MARKERS >>> REVERSE DIAGONAL')
            print('Player {} wins'.format(player))
            print(reverse_diagonal_matrix)
            
            return True
        
        elif game_board.count('X') == 5 or game_board.count('O') == 5:
            print('NO WINNING PATTERN -- GAME OVER')
            
            return True

        else:
            return False
```

```python
#---------KEEP PLAYING---------#

def keep_playing():
    yes_no = ['Y', 'N']
    keep_playing = ''

    while keep_playing not in yes_no:
        keep_playing = input('Play another time? (Y for YES, N for NO)')

        if keep_playing not in yes_no:
            print('Please select Y or N.')
    
    return True if keep_playing == 'Y' else False
```
